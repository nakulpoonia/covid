```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
```

Data Loading and Overview: Load the Covid-19 India dataset into a pandas DataFrame and display the first 5 rows to understand its structure.

```python
import pandas as pd

# Load the dataset
covid_data = pd.read_csv("Latest Covid-19 India Status.csv")

# Display the first 5 rows of the dataset
print(covid_data.head())
```

Data Comparison - Total Cases vs Population: Compare the total number of Covid-19 cases with the population size for each State/UTs using a scatter plot.

```python
# Create a scatter plot to compare total cases vs population
plt.figure(figsize=(10, 6))
plt.scatter(covid_data['Total Cases'], covid_data['Population'], color='blue', alpha=0.5)

# Add labels and title
plt.title('Total Cases vs Population for Each State/UTs')
plt.xlabel('Total Cases')
plt.ylabel('Population')

# Add State/UTs labels
for i in range(len(covid_data)):
    plt.text(covid_data['Total Cases'][i], covid_data['Population'][i], covid_data['State/UTs'][i], fontsize=8)

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

Data Visualization - Boxplot of Covid-19 Metrics: Plot a boxplot to compare the distribution of Covid-19 metrics (total cases, active cases, discharge ratio, death ratio) across different State/UTs.

```python
import seaborn as sns

# Select relevant columns for boxplot
metrics = ['Total Cases', 'Active', 'Discharge Ratio', 'Death Ratio']

# Create a boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(data=covid_data[metrics])
plt.title('Distribution of Covid-19 Metrics Across State/UTs')
plt.ylabel('Count')
plt.xlabel('Metrics')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Display the plot
plt.tight_layout()
plt.show()
```

Data Exploration - Death Ratio: Calculate the death ratio (Deaths/Total Cases) for each State/UTs and visualize it using a histogram.

```python
# Calculate death ratio
covid_data['Death Ratio'] = covid_data['Deaths'] / covid_data['Total Cases']

# Plot histogram for death ratio
plt.figure(figsize=(10, 6))
plt.hist(covid_data['Death Ratio'], bins=10, color='skyblue', edgecolor='black')

# Add labels and title
plt.title('Distribution of Death Ratio Across State/UTs')
plt.xlabel('Death Ratio')
plt.ylabel('Frequency')

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

Data Exploration - Recovery Ratio: Calculate the recovery ratio (Discharged/Total Cases) for each State/UTs and plot it on a bar chart.

```python
# Calculate recovery ratio
covid_data['Recovery Ratio'] = covid_data['Discharged'] / covid_data['Total Cases']

# Sort data by recovery ratio
sorted_data = covid_data.sort_values(by='Recovery Ratio', ascending=False)

# Plot bar chart for recovery ratio
plt.figure(figsize=(12, 8))
sns.barplot(x='Recovery Ratio', y='State/UTs', data=sorted_data, palette='viridis')

# Add labels and title
plt.title('Recovery Ratio Across State/UTs')
plt.xlabel('Recovery Ratio')
plt.ylabel('State/UTs')

# Display the plot
plt.tight_layout()
plt.show()
```

Data Exploration - Active Cases: Create a pie chart to show the distribution of active Covid-19 cases across different States/UTs.

```python
import matplotlib.pyplot as plt

# Calculate the total active cases
total_active_cases = covid_data['Active'].sum()

# Calculate the percentage of active cases for each State/UTs
covid_data['Active Percentage'] = (covid_data['Active'] / total_active_cases) * 100

# Sort the data based on the percentage of active cases
sorted_data = covid_data.sort_values(by='Active Percentage', ascending=False)

# Plot a pie chart
plt.figure(figsize=(10, 8))
plt.pie(sorted_data['Active Percentage'], labels=sorted_data['State/UTs'], autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Distribution of Active Covid-19 Cases Across State/UTs')

# Display the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

Data Analysis - Population vs Total Cases: Plot a scatter plot to analyze the relationship between population and total Covid-19 cases.

```python
# Plot scatter plot for population vs total cases
plt.figure(figsize=(10, 6))
plt.scatter(covid_data['Population'], covid_data['Total Cases'], color='orange', alpha=0.7)

# Add labels and title
plt.title('Population vs Total Covid-19 Cases for Each State/UTs')
plt.xlabel('Population')
plt.ylabel('Total Cases')

# Add State/UTs labels
for i in range(len(covid_data)):
    plt.text(covid_data['Population'][i], covid_data['Total Cases'][i], covid_data['State/UTs'][i], fontsize=8)

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

Data Analysis - Lowest Active Cases: Determine the State/UTs with the lowest number of active Covid-19 cases and display it using a horizontal bar chart.

```python
# Sort the data based on active cases in ascending order
sorted_data = covid_data.sort_values(by='Active', ascending=True)

# Plot horizontal bar chart for lowest active cases
plt.figure(figsize=(10, 8))
sns.barplot(x='Active', y='State/UTs', data=sorted_data, palette='coolwarm')

# Add labels and title
plt.title('States/UTs with the Lowest Number of Active Covid-19 Cases')
plt.xlabel('Active Cases')
plt.ylabel('State/UTs')

# Display the plot
plt.tight_layout()
plt.show()

```

Plot a pie chart illustrating the percentage of total cases, active cases, discharged cases, and deaths in India

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual Covid-19 India data)
data = {
    'Total Cases': [1000000],
    'Active': [200000],
    'Discharged': [750000],
    'Deaths': [50000]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Calculate percentages
total_cases = df['Total Cases'].iloc[0]
active_cases = df['Active'].iloc[0]
discharged_cases = df['Discharged'].iloc[0]
deaths = df['Deaths'].iloc[0]

# Calculate percentages
percentages = [
    active_cases / total_cases * 100,
    discharged_cases / total_cases * 100,
    deaths / total_cases * 100
]

# Pie chart
labels = ['Active', 'Discharged', 'Deaths']
colors = ['#ff9999','#66b3ff','#99ff99']
explode = (0.1, 0, 0)  # explode 1st slice
plt.figure(figsize=(8, 6))
plt.pie(percentages, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Covid-19 Cases in India')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```


Analyze the distribution of active cases using a violin plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have loaded your Covid-19 India data into a DataFrame called 'covid_data'

# Create a violin plot of active cases distribution
plt.figure(figsize=(12, 8))
sns.violinplot(x='Active', data=covid_data, inner='quartile')
plt.title('Distribution of Active Cases Across States/UTs in India')
plt.xlabel('Active Cases')
plt.ylabel('Density')
plt.xticks(rotation=45)
plt.show()
```

Create a bar chart showing the total number of Covid-19 cases for each State/UTs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Covid-19 India data into a DataFrame
data = {
    'State/UTs': ['Andaman and Nicobar Islands', 'Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar'],
    'Total Cases': [100, 2000, 150, 2500, 1800]
    # Add more data for other states...
}

df = pd.DataFrame(data)

# Sort the DataFrame by total cases in descending order
df_sorted = df.sort_values(by='Total Cases', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Total Cases', y='State/UTs', data=df_sorted, palette='viridis')
plt.title('Total Covid-19 Cases by State/UTs in India')
plt.xlabel('Total Cases')
plt.ylabel('State/UTs')
plt.show()
```
